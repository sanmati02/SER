```python
import os
import shutil

# Define speaker list and emotion order
speakers = ['Joey', 'Phoebe', 'Ross', 'Rachel', 'Chandler', 'Monica']
emotions = ["neutral", "joy", "sadness", "anger", "fear", "disgust", "surprise"]

# Define paths
src_dir = dev_audios          # Where the original .wav files are
dst_base_dir = '/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD_test/data/dev_splits_audio_organized'
# Make sure destination base exists
os.makedirs(dst_base_dir, exist_ok=True)

# Process each row in filtered dataframe
for _, row in train.iterrows():
    speaker = row['Speaker']
    if speaker not in speakers:
        continue
    
    emotion = row['Emotion']
    emotion_idx = emotions.index(emotion) if emotion in emotions else -1

    # Original and new file names
    original_filename = f"dia{row['Dialogue_ID']}_utt{row['Utterance_ID']}.wav"
    new_filename = f"dia{row['Dialogue_ID']}_utt{row['Utterance_ID']}_{emotion_idx}.wav"

    # Create speaker folder
    speaker_folder = os.path.join(dst_base_dir, speaker)
    os.makedirs(speaker_folder, exist_ok=True)

    # Full source and destination paths
    src_path = os.path.join(src_dir, original_filename)
    dst_path = os.path.join(speaker_folder, new_filename)

    # Move and rename the file
    if os.path.exists(src_path):
        shutil.copy2(src_path, dst_path)  # or shutil.move if you want to move instead of copy
    else:
        print(f"Missing file: {src_path}")
```

```python
import pandas as pd
import random
from collections import defaultdict
from sklearn.model_selection import train_test_split

# Load the file
file_path = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/data/test_list.txt"
with open(file_path, "r") as f:
    lines = f.readlines()

# Parse each line into (actor, emotion, full line)
data = []
for line in lines:
    line = line.strip()
    if not line:
        continue
    path, emotion = line.rsplit("\t", 1)
    actor = path.split("/")[3]  # e.g., 'Monica', 'Joey'
    data.append((actor, int(emotion), line))

# Organize data into strata (actor, emotion)
strata = defaultdict(list)
for actor, emotion, line in data:
    strata[(actor, emotion)].append(line)

# Sample 50% from each stratum
sampled_lines = []
for group, items in strata.items():
    sample_size = max(1, len(items) // 2)
    sampled_lines.extend(random.sample(items, sample_size))

# Convert to DataFrame for display
df_sampled = pd.DataFrame(sampled_lines, columns=["file_and_label"])
df_sampled
# import ace_tools as tools; tools.display_dataframe_to_user(name="Stratified 50% Sample", dataframe=df_sampled)
# Save the sampled lines to a new text file
output_path = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/data/train_fourth_list.txt"
with open(output_path, "w") as f:
    for line in df_sampled["file_and_label"]:
        f.write(f"{line}\n")

import os
import shutil
```

```python
import os
import random
import shutil
from collections import defaultdict

# === PATHS ===
input_file = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/data/test_list.txt"
base_audio_dir = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/test_splits_audio_organized"
output_list_1 = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/data/test_half1.txt"
output_list_2 = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/data/test_half2.txt"
output_dir_1 = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/test_half1"
output_dir_2 = "/Users/costanzasiniscalchi/Documents/MS/NNDL/Project/SER/MELD/test_half2"

# === READ AND PARSE FILE ===
with open(input_file, "r") as f:
    lines = [line.strip() for line in f if line.strip()]

data = []
for line in lines:
    path, emotion = line.rsplit("\t", 1)
    actor = path.split("/")[3]
    data.append((actor, int(emotion), line))

# === STRATIFIED SPLIT ===
strata = defaultdict(list)
for actor, emotion, line in data:
    strata[(actor, emotion)].append(line)

half1, half2 = [], []
random.seed(42)

for group, items in strata.items():
    items_copy = items[:]
    random.shuffle(items_copy)
    mid = len(items_copy) // 2
    half1.extend(items_copy[:mid])
    half2.extend(items_copy[mid:])

# === SAVE TEXT FILES ===
with open(output_list_1, "w") as f1:
    for line in half1:
        f1.write(f"{line}\n")
with open(output_list_2, "w") as f2:
    for line in half2:
        f2.write(f"{line}\n")

# === MOVE AUDIO FILES ===
def copy_files(file_list, dest_dir):
    os.makedirs(dest_dir, exist_ok=True)
    for line in file_list:
        rel_path = line.split("\t")[0]  # remove label
        rel_path.replace('MELD/data/test_splits_audio_organized', '')
        dst_path = os.path.join(dest_dir, rel_path)
        print(dst_path)
        os.makedirs(os.path.dirname(dst_path), exist_ok=True)
        if os.path.exists(src_path):
            shutil.copy2(src_path, dst_path)
        else:
            print(f"Missing file: {src_path}")

copy_files(half1, output_dir_1)
copy_files(half2, output_dir_2)
```
